Volteador de dígitos.PNG


Volteador de dígitos

Transformar Base 10 -> Base 2 -> Invertir -> Base 10

```python
# Input
2147483647 
1 
0
```

```python
# Output
2147483648 
4294967294 
4294967295
```

```python
def transformar(num):
    binario=[]
    exp=31
    resultado=0
    while num !=0:
        division=num//2
        residuo=num%2
        binario.append(residuo)
        num=division
    cuenta=32-len(binario)
    for i in range(cuenta):
        binario.append(0)
    inv=binario[::-1]
    for i in range(32):
        if inv[i]==0:
             inv[i]=1
        else:
            inv[i]=0
    for i in range(32):
        x=inv[i]*2**exp
        resultado+=x
        exp-=1
    return resultado
    
transformar(9)
```

```python
transformar(2147483647)
```

```python
transformar(1)
```
